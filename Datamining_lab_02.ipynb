```python
import pandas as pd
data=pd.read_csv("/content/Crime_Data_from_2020_to_Present.csv")
```

```python
summary=data.describe()
print(summary)
```

```python
print(data.describe(include=["object"]))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(data['AREA'].dropna(),bins=20,edgecolor='black',color='skyblue')
plt.title('Area distribution')
plt.xlabel('AREA')
plt.ylabel('frequency')
plt.show()
```

```python
plt.figure(figsize=(6,4))
data['AREA'].value_counts().plot(kind='bar',edgecolor='black',color='skyblue')
plt.title('Area class distribution')
plt.xlabel('AREA')
plt.ylabel('count')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.boxplot(data['AREA'].dropna(),patch_artist=True,boxprops=dict(facecolor='skyblue',edgecolor='black'))
plt.title('Box plot of AREA')
plt.ylabel('AREA')
plt.show()
```

```python
import seaborn as sns
sns.pairplot(data[['AREA','DR_NO','TIME OCC','LAT']].dropna())
plt.show()
```

```python
correlation_matrix=data[['AREA','DR_NO','TIME OCC','LAT']].corr()
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm',cbar=True)
plt.title('correlation Heatmap')
plt.show()
```

```python
sns.countplot(x='Status Desc', data=data, palette='Set2')
plt.title('Status Desc')
plt.xlabel('Status Desc')
plt.ylabel('Count')
plt.show()
```

```python
sns.violinplot(x='Status Desc',y='AREA',data=data,palette='muted')
plt.title('AREA count by status Desc Type')
plt.title('status Desc type by AREA count')
plt.show()
```
